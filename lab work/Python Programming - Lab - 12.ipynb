### 04) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt

categories = ["Engineers", "Doctors", "Business", "Others", "Artists"]
percentages = [30, 25, 20, 15, 10]
colors = ["blue", "green", "orange", "red", "purple"]

plt.pie(percentages, labels=categories, autopct="%1.1f%%", colors=colors, shadow=True)

plt.title("Pie Chart Demo")

plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt

categories = ["Engineers", "Doctors", "Business", "Teachers", "Artists"]
users_count = [120, 90, 75, 60, 50]  # Number of users in each profession

plt.bar(categories, users_count, color=['blue', 'green', 'orange', 'red', 'purple'])

plt.xlabel("Professions")
plt.ylabel("Number of Users")
plt.title("Registered Users in Matrimonial App by Profession")

plt.show()
```

### 06) WAP to demonstrate the use of Scatter Plot.

```python
import matplotlib.pyplot as plt
import random

height = [random.randint(160, 200) for _ in range(20)]
income = [random.randint(40, 100) for _ in range(20)]

plt.scatter(height, income, color='blue', marker='o', edgecolors='black')

plt.xlabel("Height (cm)")
plt.ylabel("Annual Income ($1000s)")
plt.title("Scatter Plot: Height vs Annual Income")

plt.grid(True, linestyle='--')

plt.show()
```

### 07) WAP to demonstrate the use of Histogram.

```python
import matplotlib.pyplot as plt

ages = [22, 25, 27, 24, 30, 29, 28, 33, 35, 40, 38, 22, 26, 32, 31, 
        29, 28, 34, 36, 27, 26, 30, 25, 24, 23, 32, 31, 29, 27, 26]

plt.hist(ages,  color='skyblue', edgecolor='black')

plt.xlabel('Age Groups')
plt.ylabel('Number of Users')
plt.title('Age Distribution in Matrimonial App')

plt.grid(axis='y', linestyle='--')

plt.show()
```

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt

categories = ['Engineers', 'Doctors', 'Business', 'Others', 'Artists']
values = [30, 25, 20, 15, 10]
colors = ['blue', 'green', 'orange', 'red', 'purple']

bars = plt.bar(categories, values, color=colors)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2,  
             bar.get_height(),  
             f'{bar.get_height()}%',
             ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.xlabel("Professions")
plt.ylabel("Percentage")
plt.title("Matrimonial Preferences by Profession")
plt.ylim(0, max(values) + 5)  
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import random
import matplotlib.pyplot as plt

num_points = 100
x = [random.uniform(0, 10) for _ in range(num_points)] 
y = [random.uniform(0, 10) for _ in range(num_points)]  
colors = [random.random() for _ in range(num_points)]  
sizes = [random.uniform(10, 200) for _ in range(num_points)]  

plt.figure(figsize=(8, 6))
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7, edgecolors='black')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Multiple Colors (Without NumPy)")

plt.colorbar(label="Color Intensity")

plt.show()
```
